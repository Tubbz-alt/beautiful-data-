```python
import numpy as np
import pandas as pd
import matplotlib
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
# matplotlib.rcParams['figure.figsize'] = (15, 6)
```

```python
def get_frequency_data_from_csv(data_file_path, columns = ['AF3', 'F7', 'F3', 'FC5', 'T7', 'P7', 'O1', 'O2', 'P8', 'T8', 'FC6', 'F4', 'F8', 'AF4']):
    names = ["COUNTER", "INTERPOLATED", "AF3", "F7", "F3", "FC5", "T7", "P7", "O1", "O2", "P8", "T8", "FC6", "F4", "F8", "AF4", "RAW_CQ", "GYROX", "GYROY", "MARKER", "MARKER_HARDWARE", "SYNC", "TIME_STAMP_s", "TIME_STAMP_ms"]
    data = pd.read_csv(data_file_path, header=0, names=names, dtype=np.float64)
    return pd.DataFrame(data, columns=columns)
```

```python
baseline_path = "/Users/Melisa/my-cortex-api/data/melisa/02-prepared/02-baseline.csv"
baseline_data_frame = get_frequency_data_from_csv(baseline_path)

song_path = "/Users/Melisa/my-cortex-api/data/melisa/02-prepared/03-song-01-positive.csv"
song_data_frame = get_frequency_data_from_csv(song_path)
```

```python
def plot_data_frame(data_frame):
    data_frame.plot(ylim=(3750, 5000))
```

```python
plot_data_frame(baseline_data_frame)
```

```python
plot_data_frame(song_data_frame)
```
